+ *Ara generem una nova  columna (''DepartureLate?').Mirarem si els valors per cada observació de la mostra, a la columna 
  'DepDelay' és > a 5 minuts.
  
  *Caldrà fer una petita funció que em retorni YES o NO si el DepDelay és més gran que 10 minuts, i amb el mètode "apply()"
   l'aplicarem a tots els valors de la columna*

```python
def rating_departure_delay(x):
    if x > 10.0:
        return 'YES'
    else:
        return 'NO'
```

```python
smaller_retards_df['DepartureLate?'] = smaller_retards_df['DepDelay'].apply(rating_departure_delay)
```
